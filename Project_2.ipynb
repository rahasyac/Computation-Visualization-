```python
import pandas as pd
import numpy as np
#!pip install openpyxl

# File to Load
file_to_load = "default of credit card clients.xls"

# Read Default of Credit Card Clients Dataset and store into Pandas data frame
credit_data = pd.read_excel(file_to_load)

# Changing column names to appropriate values given in dataset
headers = credit_data.iloc[0]
credit_data  = pd.DataFrame(credit_data.values[1:], columns=headers)


credit_data.head()
```

```python
#Renaming columns for better understanding 
credit_data.rename(columns={'PAY_0': 'Repayment Status - September',
                           'PAY_2': 'Repayment Status - August',
                           'PAY_3': 'Repayment Status - July',
                           'PAY_4': 'Repayment Status - June',
                           'PAY_5': 'Repayment Status - May',
                           'PAY_6': 'Repayment Status - April',
                           'BILL_AMT1': 'Bill Statement - September',
                           'BILL_AMT2': 'Bill Statement - August',
                           'BILL_AMT3': 'Bill Statement - July',
                           'BILL_AMT4': 'Bill Statement - June',
                           'BILL_AMT5': 'Bill Statement - May',
                           'BILL_AMT6': 'Bill Statement - April',
                           'PAY_AMT1': 'Amount Paid - September',
                           'PAY_AMT2': 'Amount Paid - August',
                           'PAY_AMT3': 'Amount Paid - July',
                           'PAY_AMT4': 'Amount Paid - June',
                           'PAY_AMT5': 'Amount Paid - May',
                           'PAY_AMT6': 'Amount Paid - April'}, inplace=True)

credit_data.head()
```

```python
#Changing numerical values to corresponding categorial values 
credit_data["default payment next month"].replace({1: "Defaulter", 0: "Non-defaulter"}, inplace=True)
credit_data["SEX"].replace({1: "Male", 2: "Female"}, inplace=True)
credit_data["EDUCATION"].replace({1: "Graduate school", 2: "University", 3: "High School", 4: "Others", 
                                 0:"None", 5:"PhD", 6:"Post Doctrate"}, inplace=True)
credit_data["MARRIAGE"].replace({1: "Married", 2: "Single", 3: "Others", 0: "Divorced"}, inplace=True)

credit_data.head()
```

```python
#Adding columns to get better results from the dataset 

#Bill Statement - Amount Paid per month 
credit_data["Due - September"] = credit_data["Bill Statement - September"] - credit_data["Amount Paid - September"]
credit_data["Due - August"] = credit_data["Bill Statement - August"] - credit_data["Amount Paid - August"]
credit_data["Due - July"] = credit_data["Bill Statement - July"] - credit_data["Amount Paid - July"]
credit_data["Due - June"] = credit_data["Bill Statement - June"] - credit_data["Amount Paid - June"]
credit_data["Due - May"] = credit_data["Bill Statement - May"] - credit_data["Amount Paid - May"]
credit_data["Due - April"] = credit_data["Bill Statement - April"] - credit_data["Amount Paid - April"]

#Total Bill Statement vs. Total Amount Paid 

credit_data['Total Bill']= credit_data.iloc[:,12:17].sum(axis = 1)
credit_data['Total Paid']= credit_data.iloc[:,18:23].sum(axis = 1)
credit_data['Total Due from April - September']= credit_data['Total Bill'] - credit_data['Total Paid']


credit_data.head()
#credit_data['AGE'].max()
```

```python
#Using K-means clustering to try and classify data into low, medium, and highly credible clients
from sklearn.cluster import KMeans
from sklearn import datasets
import matplotlib.pyplot as plt



# File to Load
file_k = "default of credit card clients.xls"

# Read Default of Credit Card Clients Dataset and store into Pandas data frame
credit_k = pd.read_excel(file_k)

# Changing column names to appropriate values given in dataset
headers = credit_k.iloc[0]
credit_k  = pd.DataFrame(credit_k.values[1:], columns=headers)

X = credit_k.drop(['default payment next month'], axis=1)


distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)

#To identify the optimal number of clusters we use the elbow method 
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
#From the above plot, we assume a possible number of clusters for this dataset could be = 3

kmeanModel = KMeans(n_clusters=3)
kmeanModel.fit(X)

X['k_means']= kmeanModel.predict(X)

X['k_means'].value_counts()
```

```python
from sklearn.metrics import silhouette_score


kmeanModel = KMeans(n_clusters=3)
kmeanModel.fit(X)

X["k_means"] = kmeanModel.predict(X)

s = silhouette_score(X, X["k_means"], metric='euclidean')
print (s)
```

```python
credit_data["k_means"] = kmeanModel.predict(X)
credit_data["k_means"].replace({0: "Type 1", 1: "Type 2", 2: "Type 3"}, inplace=True)
credit_data.head()
```
